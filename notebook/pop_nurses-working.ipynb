# Taking the population data from the us census bureau and filtering it down to TN and merging it to one DF

```python
pop2010_2020
```

```python
pop2010_2020_tn = pop2010_2020.loc[pop2010_2020['NAME'] == 'Tennessee']
```

```python
pop2010_2020_tn
```

```python
pop2020_2024 = pd.read_csv('../data/NST-EST2024-POPCHG2020-2024.csv')
```

```python
pop2020_2024.head()
```

```python
pop2020_2024_tn = pop2020_2024.loc[pop2020_2024['NAME'] == 'Tennessee']
```

```python
pop2020_2024_tn
```

```python
tn_pop_2010_2024 = pd.merge(pop2010_2020_tn, pop2020_2024_tn,
                             left_on=['SUMLEV', 'REGION', 'DIVISION', 'STATE', 'NAME'],
                             right_on=['SUMLEV', 'REGION', 'DIVISION', 'STATE', 'NAME'],
                             how='inner')
```

```python
tn_pop_2010_2024
```

```python
# Extract the relevant columns
columns = ['NAME', 'POPESTIMATE2017', 'POPESTIMATE2018', 'POPESTIMATE2019', 
           'POPESTIMATE2020_y', 'POPESTIMATE2021', 'POPESTIMATE2022', 
           'POPESTIMATE2023', 'POPESTIMATE2024']

# Create a subset of the DataFrame with the relevant columns
pop_estimates = tn_pop_2010_2024[columns]

# Set 'NAME' as the index (or whatever column you want on the x-axis)
pop_estimates.set_index('NAME', inplace=True)

# Plot the data for the years 2017-2024
plt.figure(figsize=(10, 6))
pop_estimates.T.plot(kind='line', marker='o', figsize=(10, 6))

# Adding labels and title
plt.title('Population Estimates for Tennessee (2017-2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population Estimate', fontsize=12)
plt.xticks(rotation=45, ha='right')
# Show the legend
plt.legend(title="Tennessee", loc='upper left')

# Display the plot
plt.grid(True)
plt.show()
```

```python
pop_estimates_melted = pop_estimates.reset_index().melt(id_vars=["NAME"], var_name="year", value_name="population")

# Ensure the 'year' column in both dataframes is of the same type (convert to string)
nurse_employment_2017_2024['year'] = nurse_employment_2017_2024['year'].astype(str)
pop_estimates_melted['year'] = pop_estimates_melted['year'].astype(str)

# Now merge the two dataframes
nurse_employment_combined = pd.merge(
    nurse_employment_2017_2024,
    pop_estimates_melted,
    left_on=['area_title', 'year'],
    right_on=['NAME', 'year'],
    how='left'
)

# Drop the 'NAME' column from pop_estimates as it's now redundant
nurse_employment_combined.drop(columns=['NAME'], inplace=True)

# Display the merged dataframe
nurse_employment_combined.head()
```

```python
nurse_employment_combined['nurses_per_capita'] = nurse_employment_combined['tot_emp'] / nurse_employment_combined['population']
```
