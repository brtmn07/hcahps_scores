# Concating all the years into 1 DF for analysis and cleaning
# adding a column to combine city and state together and making the first letter capitalized 

```python
all_star_all_years = pd.concat([all_star2017, all_star2018, all_star2019, all_star2020, all_star2021, all_star2022, 
                      all_star2023, all_star2024], axis=0)
```

```python
all_star_all_years
```

```python
all_star_all_years['city_state'] = all_star_all_years['city'] + ', ' + all_star_all_years['state']
#adding a column in to combine city and state to make it easier for presentations 
```

```python
all_star_all_years.head(1)
```

```python
# Define a function to format the city_state field
def format_city_state(value):
    if ',' in value:
        city, state = value.split(',', 1)
        city = city.strip().title()        # Capitalize first letter of each word
        state = state.strip().upper()      # Ensure state is fully uppercase
        return f"{city}, {state}"
    return value

# Apply the function to the 'city_state' column
all_star_all_years['city_state'] = all_star_all_years['city_state'].apply(format_city_state)
```

# cleaning the df for just cleanliness star rating 

```python
pd.set_option('display.max_columns', None)
```

```python
all_star_cleanliness = []
for index, row in all_star_all_years.iterrows():
    if 'Cleanliness - star rating' in row['hcahps question']:
        all_star_cleanliness.append(row)
all_star_cleanliness = pd.DataFrame(all_star_cleanliness)
```

```python
all_star_cleanliness
```

```python
all_star_cleanliness = all_star_cleanliness.loc[all_star_cleanliness['patient survey star rating'] != 'Not Available']
#removing all the not available from the df since there were not surveys for a rating 
```

```python
all_star_cleanliness
```

```python
all_star_cleanliness['patient survey star rating'] = pd.to_numeric(
    all_star_cleanliness['patient survey star rating'])
# have to change the star ratings to int
```

```python
all_star_cleanliness.to_csv('all_star_cleanliness.csv', index = False)
```

```python
all_star_cleanliness_mean = all_star_cleanliness.groupby('year')['patient survey star rating'].agg(['mean', 'count', 'std'])
```

```python
all_star_cleanliness_mean
```

```python
all_star_cleanliness_mean.plot(kind='bar', y='mean')
plt.title("AVG STAR RATING CLEANLINESS PER YEAR")
plt.xlabel("YEAR")
plt.ylabel("avg rating")
plt.grid(True)
plt.show()
```
